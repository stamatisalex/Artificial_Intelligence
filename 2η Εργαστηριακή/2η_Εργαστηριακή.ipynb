```python
# Η βιβλιοθήκη pandas είναι χρήσιμη για την εργασία με τέτοια δεδομένα
import pandas as pd 
# Διάβασμα του αρχείου 'movie_metadata.csv' 
data = pd.read_csv("movies_metadata.csv") 
#Στο csv υπαρχούν κελία με nan τιμές
#Στις θέσεις αυτές βάζουμε 'UNK' πράγμα που  κάνουμε με την παρακάτω συνάρτηση
data.fillna("UNK", inplace=True)
# Preview the first 5 lines of the loaded data 
data.head()


def clean_text(text):
  text = text.replace(u'\xa0', u'')
  text = text.replace(u"'", u'')
  return text

```

# **3ο Μέρος: Recommendation System με βάση τις Προτιμήσεις-Αξιολογήσεις του Χρήστη**

Αρχικά μελετάμε τις αξιολογήσεις κάθε χρήστη για να καταλάβουμε την δομή και τις πληροφορίες κάθε αρχείου.






**Κατασκευάζουμε τα scores για κάθε ταινία με βάση τις προηγούμενες αξιολογήσεις του χρήστη**

**Στο σημείο αυτό κατασκευάζουμε την νέα συνάρτηση η όποια θα επιστρέφει τις προτεινόμενες ταινίες συνυπολογίζοντας τις αξιολογήσεις του χρήστη.**

```python
from tqdm.notebook import tqdm
from sklearn.metrics import precision_score, recall_score, f1_score

rating_weights = {0: -1.5, 1: -1, 2:-0.5, 3:0.5, 4:1, 5:1.5}
score_weights = {1: 1, 2: 2, 3: 2.5, 4: 3.5, 5: 4, 6: 5, 7: 5.5, 8: 6.5, 9: 7, 10: 9} # ανάλογα με τα επίπεδα ομοιότητας που έχουν οριστεί στην simple_recommender
#print(score_weights)

def train_recommender(ratings, rating_weights, score_weights, start=0, end = -1, movie_score = {}):
    # Στην συνάρτηση αυτή μπορούμε να ορίζουμε ποιο υποσύνολο των αξιολογήσεων θα χρησιμοποιήσουμε για το train,
    # ενώ παράλληλα μπορούμε να αρχικοποιήσουμε και το movie_score το οποίο αποθηκεύονται τα score κάθε ταινίας.
    # Έτσι αν έχουμε κάνει train τον recommender μας για τις πρώτες 10 ταινίες, μπορούμε να συνεχίσουμε για τις 
    # υπόλοιπες χωρίς κάθε φορά να ξανά-διαμορφώνουμε το score για αυτές. Αυτό θα σας βοηθήσει στην εκτέλεση και χρονικά όταν ελέγχετε την 
    # απόδοση σε υποσύνολα των ratings π.χ. πρώτες 10 ταινίες έπειτα τις επόμενες 20 κ.ο.κ.
    if end == -1:
        end = len(ratings)

    ratings = ratings.iloc[range(start, end)]
    for row in tqdm(ratings.itertuples(index=True, name='Pandas')):
        movie = clean_text(getattr(row, 'movie_title'))
        rating = getattr(row, 'rating')

        similar_movies = simple_recommender(movie) # Βρίσκουμε όλες τις ταινίες που προτείνονται για τη συγκεκριμένη

        for similar_movie in similar_movies:  # Για κάθε τέτοια ταινία που δεν την έχει δεί ήδη ο χρήστης (άρα βαθμολογήσει)
            if similar_movie not in movie_score:
                movie_score[similar_movie] = rating_weights[int(rating)] * score_weights[movie_dict[similar_movie]]
                
            else:
                movie_score[similar_movie] += rating_weights[int(rating)] * score_weights[movie_dict[similar_movie]] #το weight θα το ορίσετε ανα επίπεδο ομοιότητας οι πολύ όμοιες ταινίες θα έχουν μεγαλύτερο βάρος
    #print(movie_score)
    return movie_score


#αυτό είναι ένα παράδειγμα για το πως θα μπορούσε να υλοποιήθει η predict
#έχουμε ορίσει ότι μια ταινία θα έπρεπε να είναι προτεινόμενη αν είχε σκορ > 0
def predict_example(ratings, movie_score):
    real, pred = [], []
    for i, row in enumerate(ratings.itertuples(index=True, name='Pandas')):
        movie = clean_text(getattr(row, 'movie_title'))
        rating = getattr(row, 'rating')

        if movie in movie_score: #αν έχουμε σχηματίσει βαθμολογία για την ταινία αυτή
            pred.append(int(movie_score[movie] > 0)) #heuristic για το αν μια ταινία είναι προτεινόμενη
            real.append(int(rating > 3))# έτσι ορίζουμε ότι μια ταινία θα έπρεπε να είναι προτεινόμενη
            #η συνθήκη αυτή δεν μπορεί να αλλάξει
        else: #δεν μπορούμε να προτείνουμε κάτι για το όποιο δεν έχουμε σχηματίσει εικόνα
            pred.append(0)
            real.append(int(rating > 3))

    return real, pred


def get_metrics(real, pred):
    metrics = {}
    metrics["precision"] = precision_score(real, pred)
    metrics["recall"] = recall_score(real, pred)
    metrics["f1"] = f1_score(real, pred)
    return metrics
```

**Επεξήγηση Κώδικα**

Έγιναν αλλαγές στα rating_weights και τα score_weights.
Όσον αφορά τα rating_weights ανατέθηκαν συμμετρικά αρνητικές και θετικές τιμές για αξιολογήσεις 2 και 3 ώστε να επηρεάζουν σε κάποιο μικρό βαθμό τις αντίστοιχες αξιολογήσεις άλλων ταινιών, καθώς θέτοντας τα 0 βλέπουμε να προκύπτουν χειρότερα αποτελέσματα.

Τα score_weights τέθηκαν ανάλογα με τα επίπεδα ομοιότητας που ορίστηκαν στην simple_recommender με μικρές διαφοροποιήσεις που εξαρτώνται από το εύρος των αλλαγών που έγιναν από το ένα επίπεδο στο άλλο και το βαθμό που αυτές τα διαφοροποιούν. Συνεπώς αν ανάμεσα σε 2 επίπεδα προστέθηκε ένα λιγότερο σημαντικό χαρακτηριστικό αυξάνουμε το score_weight κατά 0.5 ενώ για αλλαγή σημαντικότερου χαρακτηριστικού 1.

Τέλος για την εύρεση του επιπέδου ομοιότητας μιας ταινίας που χρειάζεται κατά τη κλήση της score_weights[movie_dict[similar_movie]] όπως φαίνεται χρησιμοποιείται το λεξικό movie_dict που παράγεται κατά τη κλήση του simple_recommender.

**Παρατηρήσεις**

Η εκπαίδευση του συστήματος γίνεται κλιμακωτά ανά 20-30 ταινίες καθώς σε διαφορετική περίπτωση εμφανίζεται το σφάλμα too many stacked strings που αναφέρθηκε στις ερωτήσεις της άσκησης με τον πυρήνα να μην εκτελεί το αντίστοιχο κελί.

**Συμπεράσματα**

Παρατηρώντας τις μετρικές που προκύπτουν για το σύστημα προτάσεων μας, βλέπουμε ότι με την εκπαίδευση επί του συνόλου των ταινιών επιτυγχάνει ένα βέλτιστο σκορ γύρω στο 50% για το precision και στο 60-65% για το recall με το μέσο τους να προκύπτει γύρω στο 55-57%. Τα συνολικά αυτά αποτελέσματα θεωρούνται ικανοποιητικά για ένα σύστημα με δυνατότητα προτάσεων σχετικά περιορισμένη και ένα μικρό σύνολο ταινιών προς εκπαίδευση.

Παράλληλα παρατηρείται μια γρήγορη προσέγγιση ικανοποιητικού μέσου f1 γύρω στο 50% με εκπαίδευση σε μόλις 20 από τις 100 ταινίες δείχοντας πως το σύστημα μας χρειάζεται σχετικά μικρό σύνολο εκπαίδευσης για να παρέχει ικανοποιητική απόδοση στις προτάσεις του. Ωστόσο το παραπάνω συμπέρασμα μπορεί να επηρεάζεται από το εκάστοτε υποσύνολο που παρέχεται και να οδηγεί σε καλύτερα ή χειρότερα αποτελέσματα κατά περίπτωση. Αυξάνοντας το σύνολο των ταινιών που δίνονται προς εκπαίδευση παρατηρείται μικρή βελτιώση της απόδοσης τείνοντας στις μέγιστες τιμές που αναφέρθηκαν.
